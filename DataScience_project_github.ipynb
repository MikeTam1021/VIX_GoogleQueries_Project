>First, let's pull data from the web. We will get the query data off Google and the VIX data from Yahoo.

>To do this we will import the necessary packages to effectively build a pandas DataFrame from this data.

>To give you a better understanding: Google data is pulled from Google Domestic Trends as an index set to 1 for the first day - January 1, 2004. It is a relative value of topic specific searches to total searches pegged to the initial ratio on the first day. One example, as Google describes it, would be search terms for finance and investing "stock, gold, fidelity, oil, stock market, scottrade". These are under the index 'INVEST'

>The VIX is an index calculated from near strike price options on the S&P 500 using first and second month options to produce a consistent index at 30 days out. It can be thought of as a standard deviation like metric for S&P 500 movement over a month's time to produce the most efficient options market with respect to all participant portfolios.

```python
import numpy as np
from pandas_datareader import data
import pandas as pd
from copy import deepcopy
```

>Here we will create the indices in a single list, and a root to form an iterable expression

```python
data_lis = []

root_ext = 'GOOGLEINDEX_US:'

indices = ['ADVERT', 'AIRTVL', 'AUTOBY', 'AUTOFI', 'AUTO', 'BIZIND', 'BNKRPT',
           'COMLND', 'COMPUT', 'CONSTR', 'CRCARD', 'DURBLE', 'EDUCAT',
           'INVEST', 'FINPLN', 'FURNTR', 'INSUR', 'JOBS', 'LUXURY', 'MOBILE',
           'MTGE', 'RLEST', 'RENTAL', 'SHOP', 'SMALLBIZ', 'TRAVEL', 'UNEMPL']
```

>Now, we can create a list with all the data we have pulled from the web.

>N.B. Although Google's query data goes back to January 1, 2004, pandas_datareader will only gather data beginning January 1, 2010.(At this time?) Google's query data ends abruptly at August 29, 2015 (the same as one month ago Dec 2015) - for what seems to be a simple lack of updating. As quoted on wikipedia:

>*"Originally, Google neglected updating Google Trends on a regular basis."*

>And it seems they still do. 

>For these reasons, the VIX has been pulled accordingly using a date index to determine what data google will give me. These dates get used a few more times for calculating how the dataset will be manipulated.

```python
for ind in indices:        
    data_lis.append(data.DataReader(root_ext + ind, 'google').Close)         
    data_lis[-1].name = ind

first_date = data_lis[0].index[0]
last_date = data_lis[0].index[-1]

dataVIX = data.DataReader('^VIX', 'yahoo', first_date, last_date)
dataVIX.columns = ['Open', 'High', 'Low', 'VIX', 'vol', 'adj']
data_lis.append((dataVIX.VIX))
```

>What we end up with is a list of series. So we are going to concatenate all the lists together into a single DataFrame via pandas

```python
frame_queries_VIX = pd.concat(data_lis, join= 'outer', axis = 1)
```

```python
frame_q_V_nona = frame_queries_VIX.dropna()
print frame_q_V_nona.shape
```

>## "And now for something completely different": ##

># Tamillow's Algorithm #

>I doubt it is statistically sound but I have never heard of someone doing it before. In order to prepare our dataset for regression and machine learning, we are going to sift through almost a year and a half of our correlations to find an optimal shifted matrix to use. We will also gain an understanding of some of the patterns that emerge between the VIX and our indexes. 

> So, to start off, let's create an empty list and finalize this step by filling it. We need a deep copy so references don't destroy our dataset every iteration and in future steps. Now we will iterate over a quarter of our data, and shift day by day. We will create a correlation of every value relative to the VIX, since the VIX is ultimately our label and queries are our features. Before creating this correlation we must drop the rows with missing data as we did in frame_q_V_nona. Finally, we need to take each individual value and pack it into the proper location in the list of lists, so we will have a quarter of our original dataframe lists of length 27 inside our larger list.

```python
number_of_days = last_date.date() - first_date.date()
quarter = number_of_days.days / 4

frame_shifts = deepcopy(frame_queries_VIX)

core_corr_lis = []
for shift_per in range(0, -quarter, -1):
    frame_shifts['VIX'] = frame_queries_VIX['VIX'].shift(periods=shift_per)
    corr_matrix = frame_shifts.dropna().corr(method = 'pearson')    
    core_corr_lis +=  [[corr_matrix['VIX'].loc[ind] for ind in indices]]
```

>And magically this list becomes a DataFrame through the power of pandas

```python
corr_serieses = pd.DataFrame(core_corr_lis, columns=[ind for ind in indices])

print corr_serieses.head(5)
```

>Looking closely at 'INVEST' we can see a pattern that rises in correlation between 350 and 200 days. This should indicate to us that roughly 200 days before the VIX changes, there may be some predictive value in this query. Wikipedia comments on this:

Their analysis of Google search volume for 98 terms of varying financial relevance, published in Scientific Reports, suggests that increases in search volume for financially relevant search terms tend to precede large losses in financial markets."*

>In order to check, we will test to see if the VIX is a leading indicator against finance. If the VIX is leading, then it is driving searches, if "INVEST" is leading, then that should be predicting the VIX. You may notice a high correlation for a few days as the VIX leads "INVEST" and then a sharp drop. However, looking at the data in the narrow peaks of "INVEST" gives us a confounding picture.   

>Let's check out some data!

>**N.B. The two indicators can be thought of as mirror images of one another. One is moving backwards in time, one forwards**


```python
Inv_VIX = frame_queries_VIX[['INVEST', 'VIX']]
idx = Inv_VIX['INVEST'].idxmax()

Inv_shift = deepcopy(Inv_VIX)

Inv_corr_rev_shift = []

for shift_per in range(quarter):
    Inv_shift['VIX'] = Inv_VIX['VIX'].shift(periods=shift_per)
    Inv_mat = Inv_shift.dropna().corr(method = 'pearson')
    Inv_corr_rev_shift += [Inv_mat['INVEST'].loc['VIX']]

plt.plot(Inv_corr_rev_shift, color='y', label='VIX leading')
plt.title('VIX and INVEST shifted')
plt.axhline(linewidth=1, color='r', linestyle='dashed')    
plt.plot(corr_serieses['INVEST'], color='b', label='INVEST leading')
plt.xlabel('days shifted')
plt.ylabel('Correlations')
plt.legend()
plt.show()

plt.plot(Inv_VIX['INVEST'])
plt.title('INVEST')
plt.xlabel('Time')
plt.ylabel('Index')
plt.show()

plt.plot(Inv_VIX['VIX'])
plt.title('VIX index')
plt.xlabel('Time')
plt.ylabel('Index')
plt.show()

print Inv_VIX.loc[idx-10:idx+10:1], '\n'
print Inv_VIX.tail(10)
```

>We must find the difference in the number of days between when our dataset ends and the present time to be more precise in our predictive power.

```python
from datetime import date, timedelta

now = date.today()
date = last_date.date()

difference = now - date
```

>So, now we need to find out how far to shift each feature in our dataset. We want to find the greatest correlation, positive or negative in each category. 

> Let's search for the index of the maximum of the absolute values in our corr_serieses data. And let's just find out the values associated with each.

>let's go out one month beyond the difference in our dataset days and the present, so we can try to find predictive values that may be of some use to us.

```python
template = '{0:8} {1:15} {2:4}'
print template.format('column', ' abs val corr', ' locat'),'\n'

shift_rows = []
for ind in indices:    
    i = corr_serieses[ind][corr_serieses.index > (difference.days + 31) ].abs().idxmax()
    m = corr_serieses[ind].loc[i]
    shift_rows.append(i)
    print template.format(ind, m, i)

print '\n', shift_rows
```

>#### Finally, the last step of Tamillow's Algorithm. ####

>We are going to sift through the data and move all the respective columns of the original Google Domestic Trends indexed data into new positions that will, hopefully, give us the best possible predictions. The idea is that the data is no longer aligned in order of its time series, but of its predictive value on the time series of our label, the VIX.

>After shifting everything around to it's new place, we are going to want to drop all rows of data that retain a NaN. But before that we need to reset the times so the VIX is again correctly aligned in time. So let's zip these indices and their corresponding shifts together and then unpack them to build a new dataframe. A new deepcopy would be nice as well.

>In the mean time, we are going to create a prediction dataset of features *t* period of time forward, where *t* is at minimum 31 days from the current date, and at most the minimum number of days for the maximum value in our correlation series. I felt that the easiest way to do this was chop off the end of our dataset and then manipulate it from there, thought there may have been an easier way.

>Let's do a little exploration... Looks good to me!

```python
futr_dates = [last_date + timedelta(days=i) for i in range(1,quarter+2)]

prediction_mat = frame_queries_VIX[-quarter-1:].drop(['VIX'], 1)

dataframe_shifts = deepcopy(frame_queries_VIX)

pack_sift = zip(indices, shift_rows)
for ind, st in pack_sift:
    dataframe_shifts[ind] = frame_queries_VIX[ind].shift(periods=st)
    prediction_mat[ind] = frame_queries_VIX[ind].shift(periods=-(quarter-st))

prediction_mat.index = futr_dates

dataframe_shifts = dataframe_shifts.dropna()
prediction_mat = prediction_mat.dropna()

print dataframe_shifts.tail(5), dataframe_shifts.shape, prediction_mat.head(5)
```

># End Tamillow's Algorithm

>### Now we get real 

>I would like to make predictions for one month out, the more up to date Google keeps it's Domestic Trends data, the better I would imagine the predictions to be.

>let's import sklearn and try to build something

```python
from sklearn import linear_model
```

>We should prepare our data in numpy arrays so sklearn can easily run it.

```python
X = np.array(dataframe_shifts.loc[:,:'UNEMPL']) 
y = np.array(dataframe_shifts.loc[:,'VIX'])

P = np.array(prediction_mat)
```

>And separate it into features and labels, training and test data. This will help us see if there is anything really here.

```python
tt = -30
features_train = X[:tt]
features_test = X[tt:]

targets_train = y[:tt]
targets_test = y[tt:]
```

>A linear model will fit our data best since we have a group of numerical features predicting a number. I can say, I have tested a few models and the simple linear regression model is really the best option. Ridge regression is very similar to our simple linear model since the query data is not highly correlated. Polynomial regression gives us data that has too much variance in it. One of the failings of simple linear regression is that it may not emphasize the swings in the VIX enough. Although, personally, I believe that this is not a huge failing. 

>The reality is we are trying to predict the VIX with Google query data that is delayed near a half year. This is a little unrealistic. There are hedge funds out there using a million times more computing power with millisecond data on several hundred thousand features up to the very second with hundreds of people working on them. I think their models may be much more accurate simply because they are closer to the market with all their calculations. 

>These hedgefunds are making markets much more efficiently then this notebook can. And if Google query data can not predict the collective wisdom of several thousands of institutions working simultaneously on this then so be it.

>But the fact that it may predict something that could be profitable to someone somewhere, even if misinterpreted, is nevertheless impressive. If you can see something in this data, God bless you.



```python
model_line = linear_model.LinearRegression()
model_line.fit(features_train, targets_train)

print 'Coefficients: \n', model_line.coef_

print 'Residual sum of squares: %.2f' % np.mean((model_line.predict(features_test) - targets_test) ** 2)

print 'Variance score: %.2f' % model_line.score(features_test, targets_test)

plt.plot(targets_test)
plt.title('VIX Index')
plt.xlabel('time')
plt.ylabel('index_rating')
plt.show()

plt.plot(model_line.predict(features_test), 'r')
plt.title('VIX Prediction')
plt.xlabel('time')
plt.ylabel('index_rating')
plt.show()
```

>Alright, let's take a look at the full on predictions and data we haven't included in our machine learning model. This is old VIX data that we are pulling but it is new to us and our algorithm. So the computer's prediction is a model for what the futurist Yogi Berra meant when he said "Predictions are tough, especially about the future." Some of these are past predictions. Only a month of predictions are set into the future, and we will have to see the outcome.

```python
rec_VIX = data.DataReader('^VIX', 'yahoo', futr_dates[0].date(), now)

predicts_line = pd.Series(model_line.predict(P), index=futr_dates[:min(shift_rows)+1])

plt.plot(rec_VIX['Close'], 'b')
plt.plot(predicts_line, 'r')
plt.title('Actual vs. predicted')
plt.show()

with pd.option_context('display.max_rows', 999, 'display.max_columns', 3):
    print pd.concat([rec_VIX['Close'], predicts_line], join= 'outer', axis = 1)
```
